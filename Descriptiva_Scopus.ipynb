Filas sin Abstracts se reportan como [No abstract available], entonces cuántos datos hay sin abstract

```python
df.Abstract[df['Abstract']=='[No abstract available]'].count()
```

Que porcentage del total de datos representan aquellos que no tienten Abstracts?

```python
(df.Abstract[df['Abstract']=='[No abstract available]'].count())*100/df.shape[0]
```

¿Estos Artículos sin Abstract habrán sido citados?


```python
df[['Abstract','Document Type','Cited by']][df['Abstract']=='[No abstract available]'].dropna().sort_values(by='Cited by', ascending=False)
```
